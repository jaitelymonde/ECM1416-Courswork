```python
#import packages
import numpy as np
from matplotlib.path import Path
import matplotlib.patches as patches
import matplotlib.pyplot as plt

#functions separating vertices with variable a for input
#note: part3 is separated into two sections such that it can be plotted with separate lines
def part1verts(a):
    vertices = np.matrix([
        [a, a], # P1
        [a, -a], # P2
        [-a, -a], # P3
        [-a, a],  # P4
        [a, a] # back to P1
        ])
    return vertices

def part2verts(a):
    vertices = np.matrix([
        [1-a, a-1], # P1
        [-a/4, a-1], # P2
        [-a/4, a/2], # P3
        [1-a, a/2],  # P4
        [1-a, a-1] # back to P1
        ])
    return vertices

def part3verts(a, section):
    if(section == 'a'):
        vertices = np.matrix([
            [a/4, a-1], # P1
            [a-1, a/2], # P2
        ])
        return vertices
    #then
    if(section == 'b'):        
        vertices = np.matrix([
            [a-1, a-1], # P3
            [a/4, a/2]  # P4
        ])
    return vertices

def part4verts(a):
    vertices = np.matrix([
        [a-1, -a/4], # P1
        [a-1, -a/2], # P2
        [1-a, -a/2], # P3
        [1-a, -a/4]  # P4
        ])
    return vertices

#transformation function
def verttransformation(b, verts):
    #transformation matrix
    A = np.matrix([[b, 1],[-1, 1]]) * 1/b
    
    #multiply specified vertices by the transformation matrix
    trans_verts = verts*A   

    return trans_verts

#plot original figures on graph
plt.figure()
plt.plot(part1verts(4)[:, 0], part1verts(4)[:, 1], 'ro-')
plt.plot(part2verts(4)[:, 0], part2verts(4)[:, 1], 'go-')
plt.plot(part3verts(4, 'a')[:, 0], part3verts(4, 'a')[:, 1], 'bo-')
plt.plot(part3verts(4, 'b')[:, 0], part3verts(4, 'b')[:, 1], 'bo-')
plt.plot(part4verts(4)[:, 0], part4verts(4)[:, 1], 'mo-')
plt.title('Original')
plt.xlabel('x')
plt.ylabel('y')
plt.show()

#plot transformed figures on graph
plt.figure()
plt.plot(verttransformation(3, part1verts(4))[:, 0], verttransformation(3, part1verts(4))[:, 1], 'ro-')
plt.plot(verttransformation(3, part2verts(4))[:, 0], verttransformation(3, part2verts(4))[:, 1], 'go-')
plt.plot(verttransformation(3, part3verts(4, 'a'))[:, 0], verttransformation(3, part3verts(4, 'a'))[:, 1], 'bo-')
plt.plot(verttransformation(3, part3verts(4, 'b'))[:, 0], verttransformation(3, part3verts(4, 'b'))[:, 1], 'bo-')
plt.plot(verttransformation(3, part4verts(4))[:, 0], verttransformation(3, part4verts(4))[:, 1], 'mo-')
plt.title('Transformed')
plt.xlabel('x')
plt.ylabel('y')
```

From these results, we can conclude that by utilizing a transformation matrix in multiplication with coordinates on a plane, we can manipulate those points when they are in matrix format. The results above compare the original points grouped by colour to display how each set of vertices was affected by the transformation matrix. This relates back to the theoretical foundations of computational mathematics due to the ability to multiply to matrices together causing a transformation of points.

```python
#import packages
import numpy as np
import matplotlib.pyplot as plt

# use Euler's method to solve the initial value problem numerically, 
# and compare the numerical results to the analytical results

# init_x: the initial x value
# init_y: the initial y value 
# h: the time step size
# interval: the given interval length
# steps: the number of steps within interval

# funtion to calculate initial value problem with euler's method
def initialvalueproblem(init_x, init_y, h, interval):
    steps = int (interval/h)

    t = np.linspace(0, interval, steps+1) #functions as x variable
    y = np.zeros(steps+1) #y variable
    y[0] = init_y #setting first element of y array as initial y input
    
    for n in range(0, steps):
        y[n+1] = y[n] + h*((init_x+(n*h))+(y[n]/5))
        

    #plot data    
    plt.plot(t, y, 'bo', t, (-5*t)-25 + 22/(np.exp(-1/5*t)), 'r-')    #(-5*t)-25 + 22/(np.exp(-1/5*t)) is the analytical (exact) solution
    plt.legend(['numerical', 'exact'], loc = 'upper left')
    plt.xlabel('$t$')
    plt.ylabel('$y(n)$')
    plt.show()
    
#call function
initialvalueproblem(0, -3, 1, 5)    
initialvalueproblem(0, -3, 0.2, 5)
initialvalueproblem(0, -3, 0.05, 5)
```

From these results we can concldue that by decreasing the time step size (h), we can manipulate the numerical solution (euler's method/numerical solution) to be closer to the exact, or analytical, solution. The first graph has a larger time step size of 1, decreasing to 0.2 in the second, until the last graph which has a time step size of 0.05. The final graph shows the numerical solution being closer to the exact solution than the previous two. This relates back to the theoretical foundations of computational mathematics because we are able to see what occurs when applying the euler's method to an initial value problem in comparison to the exact solution in regard to changing the time step size.
